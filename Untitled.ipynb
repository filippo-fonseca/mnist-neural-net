# MNIST Classifier (Simple ANN)

# Step 1: Setup & Loading MNIST

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# For reproducibility — so our "random" numbers are the same each run.
# Think of it like setting a specific shuffle on a deck of cards.
np.random.seed(42)
```

```python
# This downloads ~11MB. First run takes ~30 seconds, then cached.
print("Fetching MNIST... (this takes a moment the first time)")
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='liac-arff')

X = mnist.data      # the pixel data
y = mnist.target    # the labels ('0', '1', ..., '9' as strings!)

print(f"X shape: {X.shape}")   # expected: (70000, 784)
print(f"y shape: {y.shape}")   # expected: (70000,)
print(f"X dtype: {X.dtype}, y dtype: {y.dtype}")
print(f"First label: {y[0]} (type: {type(y[0])})")
```

```python
# Convert labels from strings to integers
y = y.astype(np.int64)

# Normalize pixels from [0, 255] to [0, 1]
# Why? Neural nets train MUCH better when inputs are small and centered.
# Imagine trying to tune a radio if every knob ranged from 0 to 10,000 —
# tiny turns would have huge effects. Normalizing keeps the knobs sensible.
X = X.astype(np.float32) / 255.0

print(f"X range: [{X.min():.2f}, {X.max():.2f}]")
print(f"y range: [{y.min()}, {y.max()}]")
print(f"y sample: {y[:10]}")
```

```python
fig, axes = plt.subplots(3, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    # Grab image i, reshape from (784,) back to (28, 28) for display
    img = X[i].reshape(28, 28)
    ax.imshow(img, cmap='gray')
    ax.set_title(f"Label: {y[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# The MNIST convention: first 60,000 for training, last 10,000 for testing.
# Data comes pre-shuffled in a sensible way.
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

print(f"Training set: {X_train.shape}, labels: {y_train.shape}")
print(f"Test set:     {X_test.shape}, labels: {y_test.shape}")
```

# Step 2: Preparing Data for the Network

```python
# Transpose so each COLUMN is one example
X_train = X_train.T   # shape: (784, 60000)
X_test = X_test.T     # shape: (784, 10000)

print(f"X_train shape: {X_train.shape}")   # (784, 60000)
print(f"X_test shape:  {X_test.shape}")    # (784, 10000)
```

```python
def one_hot_encode(y, num_classes=10):
    """
    Convert integer labels to one-hot encoded matrix.
    
    Input:  y of shape (m,)           e.g. [5, 0, 4, 1, ...]
    Output: Y of shape (num_classes, m)  with 1s at y[i] in column i
    """
    m = y.shape[0]
    Y = np.zeros((num_classes, m))   # start with all zeros
    Y[y, np.arange(m)] = 1           # the magic line — explained below
    return Y

Y_train = one_hot_encode(y_train)
Y_test = one_hot_encode(y_test)

print(f"Y_train shape: {Y_train.shape}")   # (10, 60000)
print(f"Y_test shape:  {Y_test.shape}")    # (10, 10000)
```

```python
# Check: pick a few examples and confirm the one-hot is correct
for i in [0, 1, 2, 100]:
    true_label = y_train[i]
    one_hot = Y_train[:, i]
    predicted_label_from_onehot = np.argmax(one_hot)
    print(f"Sample {i}: true label = {true_label}, "
          f"one-hot vector = {one_hot.astype(int)}, "
          f"argmax = {predicted_label_from_onehot} ✓" 
          if true_label == predicted_label_from_onehot else "✗")
```

# Step 3: Network Architecture & Initialization

```python
def init_params(input_size=784, hidden_size=128, output_size=10):
    """
    Initialize network parameters using He initialization.
    
    Returns a dict holding W1, b1, W2, b2.
    """
    # Layer 1: input → hidden
    # np.random.randn gives standard normal (mean 0, var 1). 
    # Multiplying by sqrt(2/n_in) gives the desired variance of 2/n_in.
    W1 = np.random.randn(hidden_size, input_size) * np.sqrt(2.0 / input_size)
    b1 = np.zeros((hidden_size, 1))
    
    # Layer 2: hidden → output
    W2 = np.random.randn(output_size, hidden_size) * np.sqrt(2.0 / hidden_size)
    b2 = np.zeros((output_size, 1))
    
    params = {"W1": W1, "b1": b1, "W2": W2, "b2": b2}
    return params

# Initialize!
params = init_params()

# Sanity checks
print(f"W1 shape: {params['W1'].shape}   expected (128, 784)")
print(f"b1 shape: {params['b1'].shape}   expected (128, 1)")
print(f"W2 shape: {params['W2'].shape}   expected (10, 128)")
print(f"b2 shape: {params['b2'].shape}   expected (10, 1)")

print(f"\nW1 stats: mean={params['W1'].mean():.4f}, std={params['W1'].std():.4f}")
print(f"W2 stats: mean={params['W2'].mean():.4f}, std={params['W2'].std():.4f}")
print(f"Expected W1 std ≈ sqrt(2/784) ≈ {np.sqrt(2/784):.4f}")
print(f"Expected W2 std ≈ sqrt(2/128) ≈ {np.sqrt(2/128):.4f}")
```

```python
# Each row of W1 is a 784-dim vector that can be reshaped to 28x28.
# It's what "image" that hidden neuron initially responds to.
# At init, these should look like noise — meaningful patterns emerge after training.
fig, axes = plt.subplots(4, 8, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    ax.imshow(params['W1'][i].reshape(28, 28), cmap='gray')
    ax.axis('off')
plt.suptitle("Initial W1 weights (32 of 128 neurons) — pure noise")
plt.tight_layout()
plt.show()
```

# Step 4: Forward Propagation

```python
def relu(Z):
    """Element-wise ReLU. Keeps positives, zeros out negatives."""
    return np.maximum(0, Z)
```

```python
test = np.array([[-2, -1, 0, 1, 2]])
print(relu(test))   # [[0 0 0 1 2]]
```

```python
def softmax(Z):
    """
    Softmax applied column-wise (each column is one example's logits).
    Includes the max-subtraction trick for numerical stability.
    """
    # Subtract the max of each column to prevent overflow
    # keepdims=True so the result has shape (1, m), which broadcasts back properly
    Z_shifted = Z - np.max(Z, axis=0, keepdims=True)
    exp_Z = np.exp(Z_shifted)
    return exp_Z / np.sum(exp_Z, axis=0, keepdims=True)
```

```python
test_logits = np.array([[1, 2, 5],
                        [2, 4, 1],
                        [3, 1, 2]])   # shape (3, 3), 3 examples, 3 classes
probs = softmax(test_logits)
print(probs)
print(f"Column sums: {probs.sum(axis=0)}")   # should all be ~1.0
```

```python
def forward_prop(X, params):
    """
    Push data X through the network.
    
    Input:
        X: (784, m) — each column is one image
        params: dict with W1, b1, W2, b2
    
    Returns:
        A2: (10, m) — predicted probabilities, each column sums to 1
        cache: dict with Z1, A1, Z2, A2 — needed for backprop later
    """
    W1, b1 = params["W1"], params["b1"]
    W2, b2 = params["W2"], params["b2"]
    
    # Layer 1: linear + ReLU
    Z1 = W1 @ X + b1          # (128, m)  — the @ operator is matrix multiply
    A1 = relu(Z1)             # (128, m)
    
    # Layer 2: linear + softmax
    Z2 = W2 @ A1 + b2         # (10, m)
    A2 = softmax(Z2)          # (10, m)
    
    cache = {"Z1": Z1, "A1": A1, "Z2": Z2, "A2": A2}
    return A2, cache
```

```python
# Run forward prop on the first 5 training examples
A2, cache = forward_prop(X_train[:, :5], params)

print(f"Output shape: {A2.shape}")   # (10, 5)
print(f"\nPredicted probabilities for first image:")
for digit, prob in enumerate(A2[:, 0]):
    bar = "█" * int(prob * 50)
    print(f"  {digit}: {prob:.4f}  {bar}")
print(f"\nSum of probabilities: {A2[:, 0].sum():.6f}   (should be 1.0)")
print(f"Predicted class: {np.argmax(A2[:, 0])}")
print(f"True label:      {y_train[0]}")
```

```python
def predict(X, params):
    """Run forward prop and return class predictions (not probabilities)."""
    A2, _ = forward_prop(X, params)
    return np.argmax(A2, axis=0)

predictions = predict(X_train, params)
accuracy = np.mean(predictions == y_train)
print(f"Untrained network accuracy: {accuracy*100:.2f}%")
print(f"(Random guessing baseline: 10%)")
```

```python
idx = 7  # pick any image
A2, _ = forward_prop(X_train[:, idx:idx+1], params)   # note the slicing to keep 2D

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Show the image
ax1.imshow(X_train[:, idx].reshape(28, 28), cmap='gray')
ax1.set_title(f"True label: {y_train[idx]}")
ax1.axis('off')

# Show the probability distribution
ax2.bar(range(10), A2.flatten())
ax2.set_xticks(range(10))
ax2.set_xlabel("Digit class")
ax2.set_ylabel("Predicted probability")
ax2.set_title(f"Network's guess: {np.argmax(A2)}")
ax2.set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
def compute_loss(A2, Y):
    """
    Cross-entropy loss averaged over the batch.
    
    Inputs:
        A2: (10, m) predicted probabilities (each column sums to 1)
        Y:  (10, m) one-hot true labels
    
    Returns:
        loss: a scalar
    """
    m = Y.shape[1]
    
    # Add a tiny epsilon inside log to prevent log(0) = -inf
    # If A2 ever contains exactly 0 for the true class, we'd get NaN.
    eps = 1e-15
    log_probs = np.log(A2 + eps)
    
    # Element-wise multiply Y * log(A2). Only the true-class entries survive
    # (because Y is 0 everywhere else).
    loss = -np.sum(Y * log_probs) / m
    
    return loss
```

```python
# Run forward prop on the full training set
A2, cache = forward_prop(X_train, params)
initial_loss = compute_loss(A2, Y_train)

print(f"Initial loss: {initial_loss:.4f}")
print(f"Expected loss for random guessing: {np.log(10):.4f}")
```

```python
def compute_accuracy(A2, y_labels):
    """
    Accuracy = fraction of examples where predicted class matches true class.
    
    Inputs:
        A2: (10, m) predicted probabilities
        y_labels: (m,) integer labels (NOT one-hot)
    """
    predictions = np.argmax(A2, axis=0)
    return np.mean(predictions == y_labels)

# Test on the training set
acc = compute_accuracy(A2, y_train)
print(f"Initial training accuracy: {acc*100:.2f}%  (expected ~10%)")
```

# Step 6: Backpropagation

```python
def relu_derivative(Z):
    """Derivative of ReLU: 1 where Z > 0, else 0."""
    return (Z > 0).astype(float)


def backward_prop(X, Y, params, cache):
    """
    Compute gradients of loss w.r.t. all parameters.
    
    Inputs:
        X:      (784, m) input batch
        Y:      (10, m)  one-hot labels
        params: dict with W1, b1, W2, b2
        cache:  dict with Z1, A1, Z2, A2 from forward prop
    
    Returns:
        grads: dict with dW1, db1, dW2, db2
    """
    m = X.shape[1]
    
    W2 = params["W2"]
    Z1, A1, A2 = cache["Z1"], cache["A1"], cache["A2"]
    
    # ===== Output layer =====
    # The magical simplification: softmax + cross-entropy → A2 - Y
    dZ2 = A2 - Y                               # (10, m)
    dW2 = (1/m) * dZ2 @ A1.T                   # (10, 128)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)  # (10, 1)
    
    # ===== Propagate gradient back through layer 2 =====
    dA1 = W2.T @ dZ2                           # (128, m)
    
    # ===== Hidden layer =====
    dZ1 = dA1 * relu_derivative(Z1)            # (128, m), element-wise multiply
    dW1 = (1/m) * dZ1 @ X.T                    # (128, 784)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)  # (128, 1)
    
    grads = {"dW1": dW1, "db1": db1, "dW2": dW2, "db2": db2}
    return grads
```

```python
def update_params(params, grads, learning_rate):
    """Apply one step of gradient descent to all parameters."""
    params["W1"] -= learning_rate * grads["dW1"]
    params["b1"] -= learning_rate * grads["db1"]
    params["W2"] -= learning_rate * grads["dW2"]
    params["b2"] -= learning_rate * grads["db2"]
    return params
```

```python
# Take a small batch to test
X_batch = X_train[:, :64]
Y_batch = Y_train[:, :64]

# Before
A2, cache = forward_prop(X_batch, params)
loss_before = compute_loss(A2, Y_batch)

# One step
grads = backward_prop(X_batch, Y_batch, params, cache)
params = update_params(params, grads, learning_rate=0.1)

# After
A2, cache = forward_prop(X_batch, params)
loss_after = compute_loss(A2, Y_batch)

print(f"Loss before: {loss_before:.4f}")
print(f"Loss after:  {loss_after:.4f}")
print(f"Decreased:   {loss_before > loss_after} ✓" if loss_before > loss_after else "Something's wrong!")
```

```python
print("Gradient shapes (should match param shapes):")
for key in ["W1", "b1", "W2", "b2"]:
    p_shape = params[key].shape
    g_shape = grads["d" + key].shape
    print(f"  {key}: param {p_shape} | grad {g_shape} | {'✓' if p_shape == g_shape else '✗'}")
```

```python
def get_mini_batches(X, Y, batch_size=64, shuffle=True):
    """
    Yield mini-batches of (X_batch, Y_batch) from the full data.
    
    Inputs:
        X: (784, m)
        Y: (10, m)
        batch_size: int
        shuffle: whether to shuffle before batching
    
    Yields:
        (X_batch, Y_batch) tuples — columns are examples
    """
    m = X.shape[1]
    indices = np.arange(m)
    
    if shuffle:
        np.random.shuffle(indices)   # in-place shuffle of index array
    
    # Walk through indices in chunks of batch_size
    for start in range(0, m, batch_size):
        end = start + batch_size
        batch_idx = indices[start:end]
        yield X[:, batch_idx], Y[:, batch_idx]
```

```python
def train(X_train, Y_train, y_train, X_test, y_test,
          hidden_size=128, learning_rate=0.1, 
          epochs=10, batch_size=64, verbose=True):
    """
    Train the neural network with mini-batch SGD.
    
    Returns:
        params: trained parameters
        history: dict with per-epoch loss and accuracy
    """
    # Initialize fresh parameters
    params = init_params(input_size=784, hidden_size=hidden_size, output_size=10)
    
    history = {
        "train_loss": [],
        "train_acc":  [],
        "test_acc":   [],
    }
    
    for epoch in range(epochs):
        epoch_loss = 0.0
        num_batches = 0
        
        # One pass through all mini-batches = one epoch
        for X_batch, Y_batch in get_mini_batches(X_train, Y_train, batch_size):
            # Forward
            A2, cache = forward_prop(X_batch, params)
            
            # Loss (for logging)
            loss = compute_loss(A2, Y_batch)
            epoch_loss += loss
            num_batches += 1
            
            # Backward
            grads = backward_prop(X_batch, Y_batch, params, cache)
            
            # Update
            params = update_params(params, grads, learning_rate)
        
        # End of epoch: compute averaged training loss, and accuracies
        avg_loss = epoch_loss / num_batches
        
        # Evaluate on full train and test sets (no updates, just measurement)
        A2_train, _ = forward_prop(X_train, params)
        train_acc = compute_accuracy(A2_train, y_train)
        
        A2_test, _ = forward_prop(X_test, params)
        test_acc = compute_accuracy(A2_test, y_test)
        
        history["train_loss"].append(avg_loss)
        history["train_acc"].append(train_acc)
        history["test_acc"].append(test_acc)
        
        if verbose:
            print(f"Epoch {epoch+1:2d}/{epochs} | "
                  f"loss: {avg_loss:.4f} | "
                  f"train acc: {train_acc*100:.2f}% | "
                  f"test acc: {test_acc*100:.2f}%")
    
    return params, history
```
