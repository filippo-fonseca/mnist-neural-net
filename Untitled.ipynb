# MNIST Classifier (Simple ANN)

# Step 1: Setup & Loading MNIST

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# For reproducibility — so our "random" numbers are the same each run.
# Think of it like setting a specific shuffle on a deck of cards.
np.random.seed(42)
```

```python
# This downloads ~11MB. First run takes ~30 seconds, then cached.
print("Fetching MNIST... (this takes a moment the first time)")
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='liac-arff')

X = mnist.data      # the pixel data
y = mnist.target    # the labels ('0', '1', ..., '9' as strings!)

print(f"X shape: {X.shape}")   # expected: (70000, 784)
print(f"y shape: {y.shape}")   # expected: (70000,)
print(f"X dtype: {X.dtype}, y dtype: {y.dtype}")
print(f"First label: {y[0]} (type: {type(y[0])})")
```

```python
# Convert labels from strings to integers
y = y.astype(np.int64)

# Normalize pixels from [0, 255] to [0, 1]
# Why? Neural nets train MUCH better when inputs are small and centered.
# Imagine trying to tune a radio if every knob ranged from 0 to 10,000 —
# tiny turns would have huge effects. Normalizing keeps the knobs sensible.
X = X.astype(np.float32) / 255.0

print(f"X range: [{X.min():.2f}, {X.max():.2f}]")
print(f"y range: [{y.min()}, {y.max()}]")
print(f"y sample: {y[:10]}")
```

```python
fig, axes = plt.subplots(3, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flat):
    # Grab image i, reshape from (784,) back to (28, 28) for display
    img = X[i].reshape(28, 28)
    ax.imshow(img, cmap='gray')
    ax.set_title(f"Label: {y[i]}")
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# The MNIST convention: first 60,000 for training, last 10,000 for testing.
# Data comes pre-shuffled in a sensible way.
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

print(f"Training set: {X_train.shape}, labels: {y_train.shape}")
print(f"Test set:     {X_test.shape}, labels: {y_test.shape}")
```

# Step 2: Preparing Data for the Network

```python
# Transpose so each COLUMN is one example
X_train = X_train.T   # shape: (784, 60000)
X_test = X_test.T     # shape: (784, 10000)

print(f"X_train shape: {X_train.shape}")   # (784, 60000)
print(f"X_test shape:  {X_test.shape}")    # (784, 10000)
```

```python
def one_hot_encode(y, num_classes=10):
    """
    Convert integer labels to one-hot encoded matrix.
    
    Input:  y of shape (m,)           e.g. [5, 0, 4, 1, ...]
    Output: Y of shape (num_classes, m)  with 1s at y[i] in column i
    """
    m = y.shape[0]
    Y = np.zeros((num_classes, m))   # start with all zeros
    Y[y, np.arange(m)] = 1           # the magic line — explained below
    return Y

Y_train = one_hot_encode(y_train)
Y_test = one_hot_encode(y_test)

print(f"Y_train shape: {Y_train.shape}")   # (10, 60000)
print(f"Y_test shape:  {Y_test.shape}")    # (10, 10000)
```

```python
# Check: pick a few examples and confirm the one-hot is correct
for i in [0, 1, 2, 100]:
    true_label = y_train[i]
    one_hot = Y_train[:, i]
    predicted_label_from_onehot = np.argmax(one_hot)
    print(f"Sample {i}: true label = {true_label}, "
          f"one-hot vector = {one_hot.astype(int)}, "
          f"argmax = {predicted_label_from_onehot} ✓" 
          if true_label == predicted_label_from_onehot else "✗")
```

# Step 3: Network Architecture & Initialization

```python
def init_params(input_size=784, hidden_size=128, output_size=10):
    """
    Initialize network parameters using He initialization.
    
    Returns a dict holding W1, b1, W2, b2.
    """
    # Layer 1: input → hidden
    # np.random.randn gives standard normal (mean 0, var 1). 
    # Multiplying by sqrt(2/n_in) gives the desired variance of 2/n_in.
    W1 = np.random.randn(hidden_size, input_size) * np.sqrt(2.0 / input_size)
    b1 = np.zeros((hidden_size, 1))
    
    # Layer 2: hidden → output
    W2 = np.random.randn(output_size, hidden_size) * np.sqrt(2.0 / hidden_size)
    b2 = np.zeros((output_size, 1))
    
    params = {"W1": W1, "b1": b1, "W2": W2, "b2": b2}
    return params

# Initialize!
params = init_params()

# Sanity checks
print(f"W1 shape: {params['W1'].shape}   expected (128, 784)")
print(f"b1 shape: {params['b1'].shape}   expected (128, 1)")
print(f"W2 shape: {params['W2'].shape}   expected (10, 128)")
print(f"b2 shape: {params['b2'].shape}   expected (10, 1)")

print(f"\nW1 stats: mean={params['W1'].mean():.4f}, std={params['W1'].std():.4f}")
print(f"W2 stats: mean={params['W2'].mean():.4f}, std={params['W2'].std():.4f}")
print(f"Expected W1 std ≈ sqrt(2/784) ≈ {np.sqrt(2/784):.4f}")
print(f"Expected W2 std ≈ sqrt(2/128) ≈ {np.sqrt(2/128):.4f}")
```
